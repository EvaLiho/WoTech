```python
# IMPORTING PANDAS LIBRARY

import pandas as pd #any other name would also work but "pd" is nad abreviation of pandas that is mainly used
```

```python
births = pd.read_csv('/content/2024_06_14_births.txt', header = None, delimiter = ' ') # SAME BUT CORRECTED COLUMNS
```

```python
births
```

```python
birth_column_names = ['Month', 'nr.of.Births']
births.columns = birth_column_names

births
```

```python
deaths = pd.read_csv('/content/2024_06_14_deaths.txt', header = None, delimiter = ' ')

death_column_names = ['Month', 'nr.of.Deaths']
deaths.columns = death_column_names

deaths
```

```python
births['Saldo'] = births['nr.of.Births'] - deaths['nr.of.Deaths']
```

```python
births['NrOfDeaths'] = deaths['nr.of.Deaths']
```

```python
births
```

### EXPORTING INTO CSV-FILE

```python
file_name = '/content/output.csv'

births.to_csv(file_name, index = False)
```
