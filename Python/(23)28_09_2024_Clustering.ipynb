### Unseperwised Training Example (Clustering)

https://scikit-learn.org/dev/api/sklearn.datasets.html

```python
from sklearn.datasets import make_blobs #crating synthetic dataset
```

```python
X, y = make_blobs(n_samples = 1000, centers = 8, cluster_std = 0.80)
```

```python
X[:,0]  # x coordinate
X[:,1]  # y coordinate
y       # the cluster of the point
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(X[:, 0], X[:, 1], c=y, edgecolors = 'k')  # y is colour, edgekolor fo edges
```

```python
from sklearn.cluster import KMeans  # import the model
```

```python
kmeans = KMeans(n_clusters = 5) # initialize the model
```

```python
kmeans.fit(X) # fir the model = train the model
```

```python
cluster_labels = kmeans.predict(X) # predict
```

```python
plt.scatter(X[:, 0], X[:, 1], c=cluster_labels, edgecolors = 'k')
```

```python
kmeans.cluster_centers_[:,0]  # cluster centers
kmeans.cluster_centers_[:,1]
```

```python
plt.scatter(X[:, 0], X[:, 1], c=cluster_labels, edgecolors = 'k')

plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], c='red', marker='X', s=200)  # adding cluster centers
```
