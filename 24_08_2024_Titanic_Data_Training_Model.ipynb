# Data IMPORT

```python
import seaborn as sns
import pandas as pd

titanic_data = sns.load_dataset('titanic')
```

```python
titanic_data.head()
```

# Data INVESTIGATION

```python
 titanic_data.dtypes   # checking datatypes
```

# Data TRANSFORMATION

```python
# modify the object categorie to string

titanic_data['class'] = titanic_data['class'].astype(str)
```

```python
titanic_data.fillna(0, inplace = True)  # replaces empty values with 0
```

```python
titanic_data.drop('deck', axis=1, inplace=True)   # removed 'deck' column because there are a lot of empty values
```

```python
titanic_data.drop(['alive', 'alone', 'adult_male', 'class'], axis=1, inplace=True)
```

```python
titanic_data.head()
```

```python
# SEX COLUMN - changing 'male', 'female' with 0 and 1

sex_mapping = {'male': 0, 'female':1}

titanic_data['sex'] = titanic_data['sex'].map(sex_mapping)
```

```python
# EMBARK TOWN COLUMN

titanic_data['embark_town'].unique()
```

```python
embark_town_mapping = {'Southampton':1, 'Cherbourg':2, 'Queenstown':3, 0:0}

titanic_data['embark_town'] = titanic_data['embark_town'].map(embark_town_mapping)
```

```python
# WHO COLUMN

titanic_data['who'].unique()
```

```python
who_mapping = {'man': 0, 'woman':1, 'child':2}

titanic_data['who'] = titanic_data['who'].map(who_mapping)
```

```python
# EMBARKED COLUMN

titanic_data['embarked'].unique()
```

```python
embarked_mapping = {'S': 1, 'C':2, 'Q':3, 0:0}

titanic_data['embarked'] = titanic_data['embarked'].map(embarked_mapping)
```

```python
titanic_data.head()
```

```python
titanic_data.info()
```

# Data MODEL

```python
from sklearn.model_selection import train_test_split
```

```python
X = titanic_data.drop('survived', axis=1)  # FEATURES
y = titanic_data['survived']              # DEPENDENT VARIABLE
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size = 0.7, random_state=42) #random_state is optional but it makes sure that every tome the split is the same
```

### Logistic Regression

```python
# IMPORT
from sklearn.linear_model import LogisticRegression
```

```python
#INITIALIZE
model = LogisticRegression()
```

```python
#TRAIN
model.fit(X_train, y_train)
```

```python
prediction_results = model.predict(X_test)

prediction_results
```
